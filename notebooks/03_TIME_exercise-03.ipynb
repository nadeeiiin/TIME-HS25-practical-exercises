```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Start with a stationary series: white noise  ~ N(0, 1)
np.random.seed(42)
n = 500
eps = np.random.normal(0, 1, n)
x = pd.Series(eps)          # stationary

# Difference at various orders
x_diff1 = x.diff().dropna()     # 1st difference
x_diff2 = x_diff1.diff().dropna()  # 2nd difference

# Compare ACFs
fig, axes = plt.subplots(3, 1, figsize=(8, 10))

plot_acf(x, lags=40, ax=axes[0])
axes[0].set_title("ACF  Original stationary series (white noise)")

plot_acf(x_diff1, lags=40, ax=axes[1])
axes[1].set_title("ACF  First difference of stationary series")

plot_acf(x_diff2, lags=40, ax=axes[2])
axes[2].set_title("ACF  Second difference of stationary series")

for ax in axes:
    ax.set_ylabel("ACF")
    ax.set_xlabel("Lag")

fig.tight_layout()
plt.show()
```
